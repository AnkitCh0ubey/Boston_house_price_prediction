```python
import pandas as pd
import numpy as np
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data1 = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
print(data1.data.shape)
```

```python
data = pd.DataFrame(data1)
print(data)
```

```python
data.head(10)
```

```python
t= ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']
titles= np.array(t)
```

```python
data.columns = titles
data
```

```python
data['Price'] = target
```

```python
data.describe()
```

```python
data.info()
```

```python
x = data
y = target
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size = 0.2,random_state = 0)
print("xtrain shape : ",xtrain.shape)
print("xtest shape : ",xtest.shape)
print("ytrain shape : ",ytrain.shape)
print("ytest shape : ",ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor =  LinearRegression()
regressor.fit(xtrain, ytrain)
y_pred = regressor.predict(xtest)
```

```python
import matplotlib.pyplot as plt
plt.scatter(ytest, y_pred, c = 'green')
plt.xlabel("True Value\n(Price in $1000's)")
plt.ylabel("Predicted value")
plt.title("True value vs Predicted value : Linear Regression")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
e_sqr = mean_squared_error(ytest, y_pred)
e_abs = mean_absolute_error(ytest,y_pred)
print("Mean Square Error : ", e_sqr)
print("Mean Absolute Error : ", e_abs)
print("Accuracy : ", 1-e_sqr) 
```
